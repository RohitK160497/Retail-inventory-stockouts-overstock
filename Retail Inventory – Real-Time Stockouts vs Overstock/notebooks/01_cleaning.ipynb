```python
import pandas as pd

# === Load Excel ===
file_path = "C:\Stevens EM\Projects\Retail Inventory – Real-Time Stockouts vs Overstock\Model Sample Data - Retail Inventory.xlsx"
# Skip 2 rows because they contain garbage/misplaced labels
df = pd.read_excel(file_path, sheet_name="Store-Product Data Sheet", skiprows=2, header=None)

# === Manually Set Column Names ===
df.columns = [
    'Store_ID', 'SKU', 'Beginning_Inventory', 'LMIL', 'EMIL', 'UMIL',
    'Sales_Units', 'Volume', 'Weight', 'Cost', 'Price',
    'Holding_Cost', 'Ordering_Cost', 'Transshipment_Cost',
    'Lead_Time', 'Min_Order_Units', 'Reorder_Units', 'Reorder_Level',
    'Unused1', 'Unused2', 'Unused3', 'Category'
]

# === Drop unused columns (at the end) ===
df = df.drop(columns=['Unused1', 'Unused2', 'Unused3'])

# === Remove any rows with missing SKUs ===
df = df[df['SKU'].notna()]

# === Save to CSV ===
df.to_csv("C:\Stevens EM\Projects\Retail Inventory – Real-Time Stockouts vs Overstock\clean_product_data.csv", index=False)

print("✅ Cleaned product data saved as 'clean_product_data.csv'")
df.head()
```
